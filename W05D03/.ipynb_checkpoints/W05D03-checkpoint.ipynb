```python
l = [(-2,2),(-1,6),(99,-2),(-3,0)]
def func(par):
    return par[-1]
sorted(l,key=func)
```

## `Map()`

```python
# syntax
map(function_name, iterable)
the map funtion returns the map object
```

```python
list_of_numbers = [2, 5, 6, 9, 3]
Output-> [4, 25, 36 , 81, 9]
```

```python
list_of_numbers = [2, 5, 6, 9, 3]
out_list = []
for num in list_of_numbers:
    out_list.append(num**2)
print(out_list)
```

```python
list_of_numbers = [2, 5, 6, 9, 3]
x = map(lambda x: x**2, list_of_numbers)
```

```python
print(x)
```

```python
print(list(x))
```

```python
list_of_num = [2,3,5,8,7]
#o/p-> [4,6,7,10,9]
```

```python
# WAP to add the number two to all the elements of the given list using map function
def add(num):
    return num+2
x = list(map(add, list_of_num))
print(x)
```

```python
l=[10.20,32,23]
x=list(map(lambda x : x+2,l))
print(x)
```

```python
list_of_fruit = ["apple","banana","mango","orange","strawberry"]
# o/p ["a","b","m","o","s"]
x = list(map(lambda s: s[0],list_of_fruit))
print(x)
```

```python
# map takes each element of the iterable and first passes it in the function as an argument to it
# then it stores the output  form that function in the same order as the form of map 
# object which is an iterable
```

```python
list_of_fruit = ["apple","banana","mango","orange","strawberry"]
# o/p ["a","b","m","o","s"]
list_of_fruit.extend(list(map(lambda s: s[0],list_of_fruit)))
print(x)
```

## `Filter`

```python
# syntax of filter
filter(function, iterable)
```

```python
bottle_of_water = ["pure water 1","impurity","pure water 2","impurity","pure water 3","impurity"]
aquaguard = list(filter(lambda drop: drop != "impurity", bottle_of_water))
print(aquaguard)
```

```python
# WAP to print the list of even numbers from the list of all numbers using filter
def evn_odd(num):
    if num%2 == 0:
        return True
    else:
        return False
list_of_numbers = [2,-2,3,0,11,141,152,189,988]
even_list = list(filter(evn_odd, list_of_numbers))
print(even_list)
```

```python
print(list(filter(lambda x: not(x%2),list_of_numbers)))
```

```python
print(list(map(lambda x: not(x%2),list_of_numbers)))
```

```python
a=[2,5,6]
y=list(map(lambda x:sum(a),a))
print(y)
print(sum(a))
```
