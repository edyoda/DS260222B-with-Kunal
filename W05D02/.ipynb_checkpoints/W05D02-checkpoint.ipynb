## `Regular Expression`

```python
Metacharacters:
    [] --> Character set -> this account for a single character
    .  --> Period -> It matches with any single character except a newline
    ^  --> Caret -> 1) It helps us to match if the string starts with particular pattern or character
                    2) If it is used inside the charecter set it is used to find all the matches  
                       except the pattern inside the character set
    $  --> Dollar -> to check if the string ends with a particular character
    *  --> star -> it matches zero or more occurence for the pattern at left side of it
    +  --> plus -> it matches one or more occurences for the pattern at the left to it
```

```python
? - Question Mark:
    The question mark symbol ? matches zero or one occurence of the pattern left to it

Regx     |     String     |    Matched
         |      gd        |     1 match
         |      god       |     1 match
         |      godisgod  |     2 match
go?d     |      good      |     0 match
```

```python
{} - Braces:
    {m,n} -> this means that at least m and at most n repetitions of pattern left to it
    {m}   -> this means that exactly m repitition of pattern
Regx     |     String     |    Matched
         |      aa        |    1 match
a{2}     |      a         |    0 match
         |      aaaa      |    2 match
         |      aadaaa    |    2 match
p{2,3}   |      pp        |    1 match
         |      ppp       |    1 match
         |      ppppp     |    2 match
         |      ppppppp   |    2 match
k{2,4}   |      kkk       |    1 match
```

```python
# findall()
    it gives you a list of all the matches
```

```python
import re
regex = "go?d"
string = "godisgod"
re.findall(regex, string)
```

```python
regex = "p{2,3}"
string = "ppppppp"
re.findall(regex,string)
```

```python
string = "ppppp"
re.findall(regex, string)
```

```python
list_of_pattern = ["aa", 
                    "a",
                    "aaaa",
                    "aadaaa"]
```

```python
for pattern in list_of_pattern:
    print(pattern,"|",
          len(re.findall("a{2}",pattern)),"matches")
```

```python
regx = '^[^A-Za-z0-9]'
list_of_pattern = ["A2031#","#254","@As58","25W"]
for pattern in list_of_pattern:
    if re.search(regx, pattern):
        print(pattern)
    else:
        print(f"{pattern} is not matched")
```
