```python
Pattern Matching
In python we have regex module i.e ('re')
```

```python
Metacharacters: A metacharacter is a character that has a special meaning during pattern processing.
[].^$*+?{}
```

```python
Pattern matching:
    password matching
    compilers
    Data scraping
    etc...
```

```python
. - Period:
    It matches any single charater except newline:
        "l"
        "@"
        "\n" not a match
[] - charater set: single character set accounts for only single character which it is containing
    [a-z] --> it will match any single lower case character
    [A-Z] --> it will match any single upper case character
    [0-9] --> it will match any single number between 0 to 9
    [2-7] --> it will match any single number between 2 to 7
    [A-Za-z] --> it will match any single alphabet
    [@M#9] --> it will be matching either of @, M or # or 9
    [A-Za-z0-9] --> it will match any single alphanumeric character
                 "asl09@" --> 5 matches
                                "a","s","l","0","9"
     [A-Za-z0-9][A-Za-z0-9] --- regx
     "asl09@"               --- test string   "as","l0" -- matches
     "rargtr"               --- test string   "ra","rg",'tr' -- matches

Q2  "rag#@%23"             --- test string  "r","a","g","#","@","%","2","3"
     regx   ---> .    
Q3   "rag#@%23" --> "ra","g#","@%","23"
     regex  ---> ..


^ : Caret Symbol:
     1) to check if the string starts with certain character
     2) to compliment or invert any charecter if used inside the character set
     
    1) the string should start with any alphabet find the regex
     "A23"  --> test string 
     A) "^[A-Za-z]"   -- regex
     "A23"   -> 1 match
     "a57"   -> 1 match
     "4a3"   -> 0 match
    2) Write a regex which matches everything but alphanumeric character
     A) "[^A-Za-z0-9]"  
     "asD5#"   -- test string
     "#"       -- 1 match
     
     3) Write a regex which ignores @ and # from the test string
     test_string = "as#A9@"
     regx = "[^@#]"
     matches = ["a","s","A","9"] --> 4 matches
     
$ - Dollar Symbol:
     It is used to check if a string ends with a certain character
     Q) Write a Regex to match a string which ends with either 9 or v
     regex = "[9v]$"
     string1 = "54vsdv"   --> 1 match
     string2 = "dksjib9"  --> 1 match
     string3 = "ksldk9va" --> 0 match
     

* - star:
     It matches zero or more occurences of the pattern at left hand side to it
     regex      |    string     | matches
                     gd         | 1 match
                     god        | 1 match
     "go*d"          good       | 1 match
                     gomod      | no match
                 goodgdgodgomd  | 3 matches
                  dooooooooog   | no match  
     ______________________________________
                |    ""         | 1 match
                |    "a"        | 1 match
     "a*"       |    "aa"       | 1 match
                |    "abcd"     | 1 match
                |    "daa"      | 1 match
                |    "aadaa"    | 2 matches
                |    "bcdhf"    | 1 match
+ - plus:
     It matches one or more occurences of the pattern at the left hand side of it
                |    ""         |  0 match
                |    "a"        |  1 match
     "a+"       |    "aa"       |  1 match
                |    "abcd"     |  1 match
                |    "daa"      |  1 match
                |    "aadaa"    |  2 matches
                |    "bcdhf"    |  0 match
```
