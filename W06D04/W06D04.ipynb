````python
class Human:
    Species = "Homosapiens"  # class attribute or class variable
    legs = 2
    university = "Edyoda"
    def __init__(self, name, gender):
        '''
        Pass name and gender as parameter eg:
        obj = Human("name", 27)
        '''
        self.name = name   # object attribute 
        self.age = 0
        self.gender = gender
        print(f"Hey Family is blessed with baby {self.gender}")
    def celebrating_birthday(self):
        '''
        This method increses the age of a person
        '''
        self.age+=1
    #Getter
    def get_current_age(self):
        '''
        This method returns us the current age of the human
        '''
        return self.age
    def get_name(self):
        '''
        This method returns us the name of the Human
        '''
        return self.name
    def get_gender(self): # self is the object reference
        '''
        This method returns the gender of the Human
        '''
        return self.gender
    
    @classmethod
    def evolution(cls, species):  #here cls is the class reference
        cls.Species = species
    
Jai = Human("Jaiprakash", "Boy")
print(Jai.get_name())
Ptksa = Human("Pratiksha", "girl")
print(Ptksa.get_name())
````

````python
class Human:
    Species = "Homosapiens"  # class attribute or class variable
    legs = 2
    university = "Edyoda"
    def __init__(self, name, gender):
        '''
        Pass name and gender as parameter eg:
        obj = Human("name", 27)
        '''
        self.name = name   # object attribute 
        self.age = 0
        self.gender = gender
        
        self.friend_list = []
        print(f"Hey Family is blessed with baby {self.gender}")
    def celebrating_birthday(self):
        '''
        This method increses the age of a person
        '''
        self.age+=1
    #Getter
    def get_current_age(self):
        '''
        This method returns us the current age of the human
        '''
        return self.age
    def get_name(self):
        '''
        This method returns us the name of the Human
        '''
        return self.name
    def get_gender(self): # self is the object reference
        '''
        This method returns the gender of the Human
        '''
        return self.gender
    
    def make_friend(self, friend): # friend is supposed to be the object ref
        self.friend_list.append(friend.name)
        friend.friend_list.append(self.name)
        print(f"{friend.name} is now friends with {self.name}")

    @classmethod
    def evolution(cls, species):  #here cls is the class reference
        cls.Species = species
    
Mayank = Human("Mayank", "Boy")
print(Mayank.get_name())
Mona = Human("Mona", "girl")
print(Mona.get_name())
````
